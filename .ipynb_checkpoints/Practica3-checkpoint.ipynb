```python
import numpy as np
```

```python
class One2Vec():
    def __init__(self, dim_in, dim_out, dim_embedding = 100):
        self.dim_in = dim_in
        self.dim_out = dim_out
        self.dim_embedding = dim_embedding
        
        self.W = np.random.random((dim_in, dim_embedding))
        self.U = np.random.random((dim_embedding, dim_out))
        
    def fit(self, X, Y, iterations = 3, learning_rate = 1):
        for i in range(iterations):
            for j, x in enumerate(X):
                salida1 = self.W[x]
                salida2 = np.dot(self.U.T, salida1)
                salida_softmax = np.exp(salida2 - np.max(salida2))/np.sum(np.exp(salida2 - np.max(salida2)))
                # error = np.array(Y[j])/np.sum(Y[j]) - salida_softmax
                error = Y[j] - salida_softmax
                
                if i < 5 or iterations - 15 < i:
                    print("i: {0}, j:{1}".format(i, j))
                    print("W:", self.W)
                    print("U.T:", self.U.T)
                    print("Salida1:", salida1)
                    print("Salida2:", salida2)
                    print("Salida_softmax:", salida_softmax)
                    print("Y[j]", Y[j])
                    print("Error:", error)
                    print("Resta1:", learning_rate * np.outer(error, salida1))
                    print("Resta2:", learning_rate*self.U.sum(1), "\n")
                self.U -= (learning_rate * np.outer(error, salida1)).T
                self.W[x] -= learning_rate * self.U.sum(1)
    
    def predict(self, x):
        salida = np.dot(self.U.T, self.W[x])
        return np.exp(salida - np.max(salida))/np.sum(np.exp(salida - np.max(salida)))        
```

```python
b = One2Vec(3,3,2)
```

```python
b.fit([0, 1, 2], [[0,0,1], [1,0,1], [1, 1, 0]], iterations = 5000, learning_rate = 1)
```

```python
b.predict(0)
```
